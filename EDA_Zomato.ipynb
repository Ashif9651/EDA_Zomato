```python
final_data=pd.merge(data,df_country,on='Country Code',how='left')
```

```python
final_data
```

```python
final_data.dtypes
```

```python
country_names = final_data.Country.value_counts().index
```

```python
country_names
```

```python
country_values = final_data.Country.value_counts().values
country_values
```

```python
final_data.Country.value_counts()
```

```python
sns.barplot(x=country_names[:5],y=country_values[:5])
plt.title('TOP 5 country with max record')
plt.show()
```

```python
final_data.columns
```

```python
ratings=final_data.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:"Rating Count"})
```

```python
ratings.head()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings,palette='tab10')
plt.show()
```

```python

c=['gray','red','Orange','yellow','green','darkgreen']
sns.barplot(x='Rating text',y='Rating Count',data=ratings,hue='Rating color',palette=c)
plt.show()
```
